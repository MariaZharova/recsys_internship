```python
import pandas as pd
from lightfm import LightFM
```

```python
clicks = pd.read_csv('./2/internship_clickstream_data.csv')
geo1 = pd.read_csv('./2/lst_announcement_data_1.gzip', compression='gzip', usecols=['id', 'geo'])
geo2 = pd.read_csv('./2/lst_announcement_data_2.gzip', compression='gzip', usecols=['id', 'geo'])
display(clicks)
display(geo1)
display(geo2)
```

```python
def get_city_id(arg):
    return eval(arg.replace('true', 'True').replace('false', 'False'))['address'][0]['id']

geo1['geo'] = geo1['geo'].apply(get_city_id)
geo2['geo'] = geo2['geo'].apply(get_city_id)
```

```python
geo1
```

```python
geo1 = geo1[geo1['geo'] == 2]
geo2 = geo2[geo2['geo'] == 2]
geo = pd.concat([geo1, geo2])
geo
```

```python
merged = clicks.merge(geo, left_on='offer_id', right_on='id', how='inner')
print(merged.shape)
merged.head()
```

```python
merged.to_csv('clicks_cleared.csv')
```
